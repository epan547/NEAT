```python
import numpy as np

# Set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from astropy.io import fits
```

```python
# Importing the image
from astropy.utils.data import download_file
image_file = download_file('https://sbnarchive.psi.edu/pds3/neat/tricam2/data/p20020815/obsdata/20020815034125c.fit.fz', cache=True )
```

```python
hdu_list = fits.open(image_file)
hdu_list.info()
```

```python
image_data = hdu_list[1].data
```

```python
print(type(image_data))
# print(image_data.shape)
print(image_data)
```

```python
# Closing FITS file, because data has been stored in variable
hdu_list.close()
```

```python
plt.figure(figsize = (30,20))
plt.imshow(image_data, cmap='gray', aspect='auto')
plt.colorbar()
```

```python
print('Min:', np.min(image_data))
print('Max:', np.max(image_data))
print('Mean:', np.mean(image_data))
print('Stdev:', np.std(image_data))
```
